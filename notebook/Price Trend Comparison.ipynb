```python
import pandas as pd

# -------------------------------
# Load dataset
# -------------------------------
df = pd.read_csv("data/data_formatted.csv", parse_dates=["posted_on"])
df["posted_on"] = pd.to_datetime(df["posted_on"], errors="coerce")

# -------------------------------
# Property Categorization
# -------------------------------
def extract_property_type(text):
    text = str(text).lower()  # handle NaN safely
    if "penthouse" in text:
        return "Penthouse"
    elif "villa" in text:
        return "Villa"
    elif "plot" in text:
        return "Plot"
    elif "studio" in text:
        return "Studio"
    elif "apartment" in text or "flat" in text:
        return "Apartment"
    else:
        return "Other"

df["property_category"] = df["about_property"].apply(extract_property_type)

# -------------------------------
# IQR Filtering Function
# -------------------------------
def remove_outliers_iqr(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return data[(data[column] >= lower) & (data[column] <= upper)]

# Apply IQR filtering on rate_sqft
df = remove_outliers_iqr(df, "rate_sqft")

# -------------------------------
# 1. RAW PROPERTY DATA (Property-level table)
# -------------------------------
raw_table = df[[
    "id", "property_category", "price_value", "area", "rate_sqft", "posted_on"
]].rename(columns={
    "id": "PropertyID",
    "property_category": "Category",
    "price_value": "Price",
    "rate_sqft": "Rate/sqft",
    "posted_on": "Posted Date"
})

raw_table.to_csv("data/raw_property.csv", index=False)

# -------------------------------
# 2. PROCESSED TIME SERIES
# -------------------------------
df["year"] = df["posted_on"].dt.year
df["month"] = df["posted_on"].dt.strftime("%b")   # Jan, Feb, ...
df["quarter"] = "Q" + df["posted_on"].dt.quarter.astype(str)

agg = df.groupby(
    ["city", "location", "year", "month", "quarter", "property_category"], 
    as_index=False
)["rate_sqft"].mean()

agg.to_csv("data/price_timeseries.csv", index=False)

# -------------------------------
# 3. Separate Trend Views
# -------------------------------
# Monthly (last 12 months)
monthly = df.groupby([df["posted_on"].dt.to_period("M")])["rate_sqft"].mean().reset_index()
monthly["period"] = monthly["posted_on"].dt.strftime("%b-%y")
monthly = monthly.sort_values("posted_on").tail(12)
monthly.to_csv("data/price_monthly.csv", index=False)

# Quarterly (last 4 quarters)
quarterly = df.groupby([df["posted_on"].dt.to_period("Q")])["rate_sqft"].mean().reset_index()
quarterly["period"] = quarterly["posted_on"].astype(str)  # e.g. 2025Q1
quarterly = quarterly.sort_values("posted_on").tail(4)
quarterly.to_csv("data/price_quarterly.csv", index=False)

# Yearly
yearly = df.groupby([df["posted_on"].dt.to_period("Y")])["rate_sqft"].mean().reset_index()
yearly["period"] = yearly["posted_on"].dt.strftime("%Y")
yearly = yearly.sort_values("posted_on")
yearly.to_csv("data/price_yearly.csv", index=False)

# -------------------------------
# 4. Master Property Type
# -------------------------------
property_master = df[["property_category"]].drop_duplicates().reset_index(drop=True)
property_master.to_csv("data/property_master.csv", index=False)
```
